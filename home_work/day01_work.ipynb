```python
# 2. 내 이미지로 YOLO11n 모델 예측하기
from ultralytics import YOLO
model = YOLO("yolo11n-pose.pt")
```

```python
results = model("tight_eyez.png",save = True) # model(), model.predict()
print(type(results)) 
print(len(results))
print(results[0])
results[0].show()
```

```python
# 3. 유튜브 URL YOLO11n 실시간 예측하기
youtube_url = "https://www.youtube.com/watch?v=kUnZ58TwS-A&t=1646s"

model.predict(source = youtube_url, show = False, save=True)
```

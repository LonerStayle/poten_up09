```python
# results = moldel(이미지,영상)
# 여러개 예측했을 때 results[이미지1, 이미지2,.....]
# 1개 예측했을 때 results [이미지1]
# 이미지1 = result[0]
```

```python
# model(), model.predict()
results = model("data/yolo_image.jpg", save = True)
print(type(results))
print(len(results))
model.predict()
```

```python
result = results[0] # 이미지 하나에 대한 결과물
```

```python
# result 에는 이미지에 대한 예측 결과 또는 정보가 담겨있는데
print(result.names) # yolo11n 모델이 찾을 수 있는 객체 딕셔너리
```

```python
# 예측 결과 확인하기 
boxes = result.boxes
boxes # cls, conf, data
# cls 예측 클래스
# conf 예측 신뢰도
# data 박스좌표, 신뢰도, 클래스
# 길이 ====> 찾은 객체의 수
```

```python
print(boxes.cls)
print(boxes.conf)
print(boxes.data)
label = result.names[int(boxes.cls[6])]   
print(label)
```

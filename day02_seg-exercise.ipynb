# 1. YOLO-seg 모델 불러오기

```python
from ultralytics import YOLO
model = YOLO("yolo11n-seg.pt")
model
```

# 2. 사진 예측하기

```python
image_url = 'https://ultralytics.com/images/bus.jpg'
results = model(image_url, save=True, project="myresult")
# results에서 이미지결과 추출하기
# 이미지결과에서 어떤 정보를 추출할 수 있는지 확인하기
```

```python
print(type(results))
print(len(results))
```

```python
result = results[0]
result
```

```python
names = result.names
print(names)
```

```python
boxes = result.boxes
boxes
```

```python
masks = result.masks
print(len(masks.data))
masks.data
```

```python
# 이미지 출력하기
import matplotlib.pyplot as plt 
mask = masks.data[5].detach().cpu().numpy()
plt.imshow(mask)
plt.axis("off")
plt.show()
```

# 3. FastSAM 모델 사용하기

#### 1) 모델 불러오기

```python
from ultralytics import FastSAM
model = FastSAM("FastSAM-s.pt")
model
```

#### 2) 이미지 예측하기

```python
source = "dog.jpeg"

everything_results = model(source, save=True, project="myresult", conf=0.7)
len(everything_results)
```

```python
result = everything_results[0]
result
```

```python
masks = result.masks
```

```python
# 이미지 출력하기
import matplotlib.pyplot as plt 
mask = masks.data[5].detach().cpu().numpy()
plt.imshow(mask)
plt.axis("off")
plt.show()
```

#### 3) 이미지와 텍스트로 예측하기

```python
results = model(source, texts="a photo of a cap", save=True, project="myresult")
```

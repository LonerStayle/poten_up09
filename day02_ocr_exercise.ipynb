# 2. OCR로 이미지에서 텍스트 추출하기

```python
from paddleocr import PaddleOCR
ocr = PaddleOCR(
    use_doc_orientation_classify=False,
    use_doc_unwarping=False,
    use_textline_orientation=False
)

results = ocr.predict(
    input = "https://paddle-model-ecology.bj.bcebos.com/paddlex/imgs/demo_image/general_ocr_002.png"
)

for res in results:
    res.print()
    res.save_to_img("output")
```

```python
from paddleocr import PaddleOCR
ocr = PaddleOCR(
    lang="korean",
    use_doc_orientation_classify=False,
    use_doc_unwarping=False,
    use_textline_orientation=False
)
image_path = "webtoon1.jpeg"
resultss = ocr.predict(input=image_path)
```

```python
result0 = resultss[0]
result0
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "webtoon1.jpeg"
image = Image.open(image_path)
plt.figure(figsize=(3,3))
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
from PIL import ImageDraw, ImageFont
image_news = image.copy()
draw = ImageDraw.Draw(image_news)
font_path="C:Windows/Fonts/malgun.ttf"
font = ImageFont.truetype(font_path,16)
```

```python
#Box 생김새 확인하기
result0["rec_boxes"][0].tolist()
```

```python
my_text_list = ["오늘 서영님 생일이래","대박이지?","생일날에도 코딩한대 ~"]
```

```python
for idx, (x1,y1,x2,y2) in enumerate(result0["rec_boxes"]):
    print(x1,y1,x2,y2)
    draw.rectangle([x1,y1,x2,y2], outline="white", fill="white") 
    draw.text([x1,y1,x2,y2], my_text_list[idx], font = font,fill="black")
```

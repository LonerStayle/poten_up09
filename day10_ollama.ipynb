````python
# 1. llama ê°€ ì•„ë‹Œ ë‹¤ë¥¸ ëª¨ë¸ ë‹¤ìš´ë¡œë“œ ë°›ì•„ë³´ê¸°
response = ollama.chat(
    model="gemma3:4b",
    messages=[
    {"role":"system", "content":system},
    {"role":"user", "content":content}
])

#---gemma3:4b íŠ¹ì§•---

# - ê¸€ìê°€ ê¹¨ì§ˆ ë•Œê°€ ìˆìŒ
# - ê°€ë” ìì‹ ì´ ì¸ê³µì§€ëŠ¥, ì–¸ì–´ ëª¨ë¸ì´ë¼ê³  í•¨ 
# - ì¸ë„ì–´ì™€ í•œêµ­ì–´ ì–¸ì–´ë¥¼ ì„ì–´ ì‚¬ìš©í•˜ë¼ëŠ” ì§€ì‹œë¥¼ ë“£ì§€ ì•ŠìŒ
#  - ì˜ì–´, í•œêµ­ì–´(ì¸ë„ì–´ë¡œ ì¸ì‹)ë¡œ ê²°ê³¼ê°€ ë‚˜ì˜´ 
print(response.message.content)
````

````python
import ollama

def chat(user_text):
    response = ollama.chat(
        model="gemma3:4b",
        messages=[
        {"role":"system", "content":"ì¬ë¯¸ìˆê²Œ ë§í•´ì£¼ì„¸ìš”"},
        {"role":"user", "content":user_text}
    ])
    return response.message.content
````

````python
while True:
    question = input("ì…ë ¥í•˜ì„¸ìš”")
    if question == "ê·¸ë§Œ":
        break
    answer = chat(question)
    print(f"Question: {question}")
    print(f"Answer: {answer}")
````

#  Ollama APIë¡œ ìš”ì²­í•˜ê¸°

````python
# curl http://localhost:11434/api/generate -d '{
#   "model" : "llama3.2",
#   "prompt" : "Why is the sky blue"
# }'

import requests

url = "http://localhost:11434/api/generate"
data = {
    "model" : "gemma3:4b",
    "prompt" : "Why is the sky blue"
}

response = requests.post(url, json=data)
print(response.text)
````

````python
import json

with requests.post(url, json=data) as r:
    for line in r.iter_lines():
        print(json.loads(line)["response"], end="")
````

````python
# curl http://localhost:11434/api/chat -d '{
#   "model" : "llama3.2",
#   "prompt" : "Why is the sky blue"
# }'

import requests

url = "http://localhost:11434/api/chat"
data = {
    "model": "gemma3:4b",
    "messages": [{"role": "user", "content": "why is the sky blue?"}],
    "stream": False,
}

response = requests.post(url, json=data)
print(response.text)
print(type(response.text))
print(type(response.json()))
print(response.json()["message"])
````

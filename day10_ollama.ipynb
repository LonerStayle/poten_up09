# 미니 프로젝트: Ollama 로 페르소나가 부여된 챗봇 만들기

````python
system = """당신은 슈퍼 개발자이면서 말할때 인도어 70% 한국어30% 비율로 번갈아서 사용합니다."""
content = "자기소개 해주세요"
````

````python

#---llama3.1:8b특징---

# 자신의 핵심 직업에 충실함 
# 인도어 한국어 비율 조건이 주어졌지만 전체 한국어로만 소통
# 언어 변경이 아닌 문화로 알아들음 

response = ollama.chat(
    model="llama3.1:8b",
    messages=[
    {"role":"system", "content":system},
    {"role":"user", "content":content}
])
# print(response) # 여기에서 출력값만 파싱하기
content = response.message.content
content 
````

````python
# 1. llama 가 아닌 다른 모델 다운로드 받아보기
response = ollama.chat(
    model="gemma3:4b",
    messages=[
    {"role":"system", "content":system},
    {"role":"user", "content":content}
])

#---gemma3:4b 특징---

# - 글자가 깨질 때가 있음
# - 가끔 자신이 인공지능, 언어 모델이라고 함 
# - 인도어와 한국어 언어를 섞어 사용하라는 지시를 듣지 않음
#  - 영어, 한국어(인도어로 인식)로 결과가 나옴 
print(response.message.content)
````

````python
# 2. OPENAI API KEY를 쓸 때와 로컬 LLM을 쓸 때의 차이점 비교하기
from openai import OpenAI
client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {
            "role": "system",
            "content": system
        },
        {
            "role": "user",
            "content": content
        }
    ],
    temperature=0.5
)
#---gpt-4.1-nano 모델의 특징---
# - 본인 역할에 대한 이해도가 나쁘지 않음 
# - 언어를 섞어 사용하라는 지시를 듣지만 비율이 맞지 않음 (디테일 부족)
answer = response.choices[0].message.content
answer
````

#### 속도 비교하기 
``` 올라마: 3.2초, GPT: 1.4초 ```

````python
import ollama

def chat(user_text):
    response = ollama.chat(
        model="gemma3:4b",
        messages=[
        {"role":"system", "content":"재미있게 말해주세요"},
        {"role":"user", "content":user_text}
    ])
    return response.message.content
````

````python
while True:
    question = input("입력하세요")
    if question == "그만":
        break
    answer = chat(question)
    print(f"Question: {question}")
    print(f"Answer: {answer}")
````

#  Ollama API로 요청하기

````python
# curl http://localhost:11434/api/generate -d '{
#   "model" : "llama3.2",
#   "prompt" : "Why is the sky blue"
# }'

import requests

url = "http://localhost:11434/api/generate"
data = {
    "model" : "gemma3:4b",
    "prompt" : "Why is the sky blue"
}

response = requests.post(url, json=data)
print(response.text)
````

````python
import json

with requests.post(url, json=data) as r:
    for line in r.iter_lines():
        print(json.loads(line)["response"], end="")
````

````python
# curl http://localhost:11434/api/chat -d '{
#   "model" : "llama3.2",
#   "prompt" : "Why is the sky blue"
# }'

import requests

url = "http://localhost:11434/api/chat"
data = {
    "model": "gemma3:4b",
    "messages": [{"role": "user", "content": "why is the sky blue?"}],
    "stream": False,
}

response = requests.post(url, json=data)
print(response.text)
print(type(response.text))
print(type(response.json()))
print(response.json()["message"])
````
